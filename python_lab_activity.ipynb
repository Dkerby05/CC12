```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

sb.set_style('whitegrid')
```

```python
mean_1 = 4300
std_dev_1 = 750
x1_1 = 2500
x2_1 = 4200

# Probability computation
prob1_lower = norm(loc=mean_1, scale=std_dev_1).cdf(x1_1)
prob1_upper = norm(loc=mean_1, scale=std_dev_1).cdf(x2_1)
prob1_between = prob1_upper - prob1_lower
print(f"1. P(2500 < x < 4200) = {prob1_between:.4f} or {prob1_between:.2%}")

# Visualization
data_1 = np.arange(1000, 7000, 1)
pdf_1 = norm.pdf(data_1, loc=mean_1, scale=std_dev_1)
sb.lineplot(x=data_1, y=pdf_1, color='black')
plt.xlabel('Acres Burned')
plt.ylabel('Probability Density')
plt.title('Forest Fire Distribution (2500 to 4200 acres)')
plt.axvline(x1_1, color='blue', linestyle='--')
plt.axvline(x2_1, color='blue', linestyle='--')
plt.fill_between(data_1, pdf_1, where=(data_1 >= x1_1) & (data_1 <= x2_1), color='red', alpha=0.5)
plt.show()
```

```python
mean_2 = 27989
std_dev_2 = 3250
x1_2 = 20000
x2_2 = 30000

# Probability computation
prob2_lower = norm(loc=mean_2, scale=std_dev_2).cdf(x1_2)
prob2_upper = norm(loc=mean_2, scale=std_dev_2).cdf(x2_2)
prob2_between = prob2_upper - prob2_lower
print(f"2. P(20000 < x < 30000) = {prob2_between:.4f} or {prob2_between:.2%}")

# Visualization
data_2 = np.arange(15000, 40000, 10)
pdf_2 = norm.pdf(data_2, loc=mean_2, scale=std_dev_2)
sb.lineplot(x=data_2, y=pdf_2, color='black')
plt.xlabel('Salary (PHP)')
plt.ylabel('Probability Density')
plt.title('Teacher Salary Distribution (20,000 to 30,000 PHP)')
plt.axvline(x1_2, color='blue', linestyle='--')
plt.axvline(x2_2, color='blue', linestyle='--')
plt.fill_between(data_2, pdf_2, where=(data_2 >= x1_2) & (data_2 <= x2_2), color='red', alpha=0.5)
plt.show()
```
